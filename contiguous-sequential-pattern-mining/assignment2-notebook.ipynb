```python
from itertools import combinations, permutations
import numpy as np
import time
```

```python
# create the reviews lists
review_lists = []
with open('reviews.txt', 'r') as f:

    for line in f.readlines():

        l = line.strip().split(' ')
        review_lists.append(l)

f.close()
```

```python
RELATIVE_SUPPORT = 0.01
MIN_ABS_SUPPORT = int(RELATIVE_SUPPORT * len(review_lists))
print(MIN_ABS_SUPPORT)
```

```python
# map each element to an integer id
word_ids = {}
word_id = 0

for lst in review_lists:

    for word in lst:

        if word not in word_ids:

            word_ids[word] = word_id
            word_id = word_id + 1
```

```python
id_review_lists = []

for lst in review_lists:

    id_review_list = []
    
    for word in lst:

        id = word_ids[word]
        id_review_list.append(id)

    id_review_lists.append(id_review_list)
```

```python
# find the number of frequent 1 patterns by counting the number
# of times each word appears in the reviews lists
with open('patterns.txt', 'w') as f:

    # initialize the count of number of reviews that a word appears in
    word_id_counts = {}
    frequent_word_ids = []

    # perform the count
    for lst in id_review_lists:

        for word_id in lst:

            if word_id not in word_id_counts:

                word_id_counts[word_id] = 0

            word_id_counts[word_id] = word_id_counts[word_id] + 1

    # write the frequent one sets to the output dict and add the
    # frequent ids to the list of ids
    for word_id in word_id_counts.keys():

        count = word_id_counts.get(word_id)

        if count > MIN_ABS_SUPPORT:

            word = list(word_ids.keys())[list(word_ids.values()).index(word_id)]

            # write line to the output file
            line = f"{count}:{word}\n"
            f.write(line)

            # append word id to ids of the 
            frequent_word_ids.append(word_id)
```

```python
length = 2

prev_freq_patterns = frequent_word_ids

with open('patterns.txt', 'a') as f:

    while length < len(frequent_word_ids):

        curr_freq_patterns = []

        # iterate through the reviews and generate all length-k
        # contiguous patterns

        length_k_phrases = []
        
        for lst in id_review_lists:

            for i in range(len(lst) - (length - 1)):

                phrase = tuple(lst[i:(i+length)])
                
                length_k_phrases.append(phrase)

        num_phrases = len(set(length_k_phrases))

        # count the number of appearances of pruned phrases
        for i, phrase in enumerate(set(length_k_phrases)):

            print(f"Phrase {i} of {num_phrases}")

            # prune each phrase
            phrase_subsets = combinations(phrase, length - 1)
            for subset in phrase_subsets:
                if subset not in prev_freq_patterns:
                    break

            count = length_k_phrases.count(phrase)

            if count > MIN_ABS_SUPPORT:

                words = ''

                for i, id in enumerate(phrase):

                    word = list(word_ids.keys())[list(word_ids.values()).index(id)]

                    if i != len(phrase) - 1:

                        words = words + word + ";"

                    else:

                        words = words + word
                    
                line = f"{count}:{words}\n"
                f.write(line)

        if not curr_freq_patterns:
            break

        length = length + 1

        print(f"Now discovering length {length} phrases...\n")
```
